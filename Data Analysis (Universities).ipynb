# 1st Question
Do universities with a high number of applications are the preferred ones by students? in other words, could the number of applications tell us that a university is one of the most preferred by students?

```python
#lets see which universisties got higher applications(top20)

high_app = cleaned_unidata[['Name','Applicants total']].sort_values("Applicants total", ascending=False).head(20)
high_app
```

```python
plt.figure(figsize=(12,12))
sns.barplot(x='Applicants total',y='Name',data=high_app )

plt.title('Top 20 American Universities with the Most Applications in 2013')

plt.xlabel('Number of applications')
```

We could think that the universities with the highest number of applications are the preferred ones by students. However, one student can send several applications at the same time, but only one university could be the preferred one.

```python
plt.figure(figsize=(16,6))

plt.subplot(1,3,1)
sns.histplot(cleaned_unidata['Applicants total'],bins=50)
plt.title('Histogram of Number of Applications. ')
plt.xlabel('Number of Applications')
plt.axis([0,30000,0,350])
plt.xticks(rotation=45)
plt.grid()

plt.subplot(1,3,2)
sns.histplot(cleaned_unidata['Admissions total'],bins=50)
plt.title('Histogram of Number of Admissions.')
plt.axis([0,10000,0,350])
plt.xlabel('Number of Admissions')
plt.xticks(rotation=45)
plt.grid()

plt.subplot(1,3,3)
sns.histplot(cleaned_unidata['Enrolled total'],bins=50)
plt.title('Histogram of Number of Enrollments.')
plt.axis([0,5000,0,350])
plt.xlabel('Number of Enrollments')
plt.grid()
plt.xticks(rotation=45)
plt.tight_layout(pad=2)
```

As we can see, the majority of universities received less than 5000 applications. However, some universities received a much higher number of applications, as many as 30000 applications in the same year.

Talking about enrollments, we'll find lower numbers when compared to the number of applications and admissions. That happens because students usually apply to several universities at the same time, hoping that the favorite one admits them; however, when that is not the case, they still have other options.

So we can conclude, highest number of applications need not mean students are interested in it. 

# 2nd Question
Do students prefer universities that have a high rate of admission? in other words, do students prefer a university where it is easier for them to be admitted?

```python
cleaned_unidata['acceptance_rate'] =(cleaned_unidata['Admissions total']/cleaned_unidata['Applicants total']* 100) 
cleaned_unidata['enrollment_rate'] =(cleaned_unidata['Enrolled total']/cleaned_unidata['Admissions total']* 100) 
```

```python
plt.figure(figsize = (8,8))
sns.scatterplot(x ='Applicants total' , y = 'enrollment_rate' , data = cleaned_unidata)
plt.xlabel('Number of Applications')
plt.ylabel('Enrollment rate %')
plt.grid()
```

We can see that,majority of lower number of applications have higher percentage of enrollment rate. There are few exceptions though. This pattern indicates that high numbers of applications do not equate to higher student preference for a university

# 5th Question
Do students prefer a university for its low on-campus cost of living?

```python
#lets do it for in state students and out students separately

plt.figure(figsize=(10,7))

plt.subplot(1,2,1)

sns.barplot(x=cleaned_unidata['Control of institution'], y=cleaned_unidata['Total price for in-state students living on campus 2013-14'])
plt.title('''Average and variation of the Cost for 
In-State Students Living on Campus (2013-2014)''', fontsize=10)        
plt.ylabel('Cost of living on campus $')
plt.show()

plt.figure(figsize=(10,7))
plt.subplot(1,2,2)

sns.scatterplot(x=cleaned_unidata['Total price for in-state students living on campus 2013-14'],y=cleaned_unidata['enrollment_rate'],hue=cleaned_unidata['Control of institution']);
plt.title('''Cost for In-State Students Living  
on Campus vs Enrollment Rate (2013-2014)''', fontsize=10)
plt.xlabel('Cost of living on campus $')
plt.ylabel('Enrollment Rate')
plt.xticks(rotation=45)

plt.tight_layout(pad=2)
```

```python
plt.figure(figsize=(16,7))


plt.subplot(1,2,1)
sns.barplot(x=cleaned_unidata['Control of institution'], y=cleaned_unidata['Total price for out-of-state students living on campus 2013-14'])
plt.title('''Average and variation of the Cost 
for Out-State Students Living on Campus (2013-2014)''')       
plt.ylabel('Cost of living on campus $')

plt.subplot(1,2,2)
sns.scatterplot(x=cleaned_unidata['Total price for out-of-state students living on campus 2013-14'],y=cleaned_unidata['enrollment_rate'], hue=cleaned_unidata['Control of institution']);
plt.title('''Cost for Out-State Students Living 
on Campus vs Enrollment Rate (2013-2014)''')
plt.xlabel('Cost of living on campus $')
plt.ylabel('Enrollment Rate')
plt.tight_layout(pad=2)
```

From Two graphs, it is apparent that public universities offer much more affordable options for in-state students as well as out-state students compared to private universities, with an average cost less than private universities.

It is also observed a trend of higher enrollment rates at universities with lower tuition and fees. This suggests that affordability plays a significant role in students' college choices.
