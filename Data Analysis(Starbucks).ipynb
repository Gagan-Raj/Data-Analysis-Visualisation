```python
import matplotlib
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
test_ds = pd.read_csv('Data/starbucks_drinkMenu_expanded.csv')

test_ds.rename(columns={'Calories':'calories'},inplace=True)
test_ds.head()
```

```python
test_ds.columns
```

```python
test_a = pd.read_csv('Data/starbucks-menu-nutrition-drinks.csv')
test_a.rename(columns={'Unnamed: 0':'name','Calories':'calories','Fat (g)':'fat','Carb. (g)':'carb','Fiber (g)':'fiber','Protein':'protein',
                      'Sodium':'sodium'},inplace=True)
test_a.head()
```

```python
test_a.columns
```

```python
test_b = pd.read_csv('Data/starbucks-menu-nutrition-food.csv',encoding="utf-16")

test_b.rename(columns={'Unnamed: 0':'name',' Calories':'calories',' Fat (g)':'fat',' Carb. (g)':'carb',' Fiber (g)':'fiber',' Protein (g)':'protein'
                     },inplace=True)
test_b.head()
```

```python
test_b.columns
```

```python
test_data=pd.concat([test_a,test_b],ignore_index=True,sort=False)

test_data.head()
```

```python
test_ds = test_ds.rename(columns={' Total Fat (g)': 'Total Fat', 'Trans Fat (g) ': 'Trans Fat',
                                            'Saturated Fat (g)': 'Saturated Fat', ' Sodium (mg)': 'Sodium',
                                            ' Total Carbohydrates (g) ': 'Total Carbohydrates',
                                            'Cholesterol (mg)': 'Cholesterol', ' Dietary Fibre (g)': 'Dietary Fibre',
                                            ' Sugars (g)': 'Sugars', ' Protein (g) ': 'Protein',
                                            'Vitamin A (% DV) ': 'Vitamin A (% DV)', ' Calcium (% DV) ': 'Calcium (% DV)',
                                            'Iron (% DV) ': 'Iron (% DV)', 'Caffeine (mg)': 'Caffeine'})
```

```python
test_ds = test_ds.replace({'Vitamin A (% DV)': {'%': ''}, 'Vitamin C (% DV)': {'%': ''},
                                     'Calcium (% DV)': {'%': ''}, 'Iron (% DV)': {'%': ''}}, regex=True)
```

```python
test_ds.isnull().sum()
```

```python
test_ds.describe()
```

```python
test_g=pd.merge(test_ds,test_data,on='calories',how='inner')
test_g.to_csv('nutri.csv',index=False)
test_g.head()
```

```python
test_g.columns
```

```python
test_g.loc[249, 'Total Fat'] = '3.2'
test_g.loc[250, 'Total Fat'] = '3.2'
test_g = test_g.replace({'Caffeine': {'Varies': None, 'varies': None}})
test_g = test_g.replace({'Iron (% DV)': {'.00': ''}}, regex=True)
test_g.head()
```

```python
# We use the .apply function to apply .to_numeric to multiple columns
columns_to_change = ['Total Fat', 'Caffeine', 'Vitamin A (% DV)', 'Vitamin C (% DV)', 'Calcium (% DV)', 'Iron (% DV)']
test_g[columns_to_change] = test_g[columns_to_change].apply(pd.to_numeric)

test_g.replace('-', '', regex=True, inplace=True)
columns_to_change = ['calories', 'fat', 'carb', 'fiber', 'Protein', 'Sodium']
test_g[columns_to_change] = test_g[columns_to_change].apply(pd.to_numeric)

```

```python
plt.figure(figsize=(12, 7))

mask = np.triu(np.ones_like(test_g.corr(), dtype=bool))

corrplot = sns.heatmap(test_g.corr(), vmin=-1, vmax=1, annot=True, mask=mask, cmap='Spectral').set_title("Correlation Plot of all columns") 

plt.show()
```

We see a strong corelation between Sugar-Cholesterol & Sodium-Saturated Fat-Trans Fat & Cholesterol- Total Carbohydrates & Sugars-Total Carbohydrates.

```python
test_g['Caffeine'].info()
```

```python
plt.figure(figsize=(15,7))
sns.set_theme(style='whitegrid')
sns.boxplot(data=test_g, x='Beverage_category', y='Caffeine', palette='hls').set(title='Distribution of Caffeine by Beverage Prep')
plt.xticks(rotation=45)
plt.show()
```

We can see the distribution of caffeine in each beverage. Frappuccino blended coffee has highest average value, wjile frappuccino blended creme has lowest

```python
plt.figure(figsize=(7,4))
plot = sns.barplot(y=test_g['calories'], x=test_g['Beverage_category']).set(xlabel='Category', ylabel='Average Calories', title='Average Calories per Beverage Category')
plt.xticks(rotation=90)
plt.show()
```

Frappuccino blended coffee and smoothies has higher average calorie, folllowed by signature expresso drinks. There may also be link to highest caffine to higher calories

```python
plt.figure(figsize=(5, 5))
sns.set_theme(style='whitegrid')
sns.scatterplot(data=test_g, x='calories', y='fat', hue='carb', palette='Spectral').set(xlabel='Calories', ylabel='Fat (g)', title='Fat Content versus Calories')
plt.show()
```

We can observe that on an average, lower Fat and higher carb has higher calories when compared to the other values

what food has the most amount of calories?

```python
#considering only food nutrition data given already
top20_cal = test_b.sort_values('calories', ascending=False).head(20)
top20_cal
```

```python
plt.figure(figsize=(8, 6))
sns.set_theme(style='whitegrid')
plot = sns.barplot(data=top20_cal, x='name', y='calories', palette='rainbow')
plot.set(title='Highest Calorie Foods', xlabel='Food Item')
plt.xticks(rotation=90)
plt.show()
```

We can infer that lentils & vegetable protein bowl with brown rice has highest calories, while homestyle chicken has lowest
